```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculating total number of schools with .nunique
total_schools = school_data_complete["school_name"].nunique()
total_schools
```

```python
#calculating the total number of students wit hteh student ID and .count
total_students = school_data_complete["Student ID"].count()
total_students
```

```python
#Calculating the total budget, by first figuring out the unique budget counts to the dataset.
budget_unique = school_data_complete["budget"].unique()
budget_unique
```

```python
#Adding up the unique budgets from previous cell will give us the total budget
total_budget = budget_unique.sum()
total_budget
```

```python
#checking the data type of the maths score to see if it's in integer
school_data_complete["maths_score"].dtypes
```

```python
#calculating the average with mean function on the maths-score column
average_maths_score = school_data_complete["maths_score"].mean()
average_maths_score
```

```python
#calculating the average with mean function on the reading_score column
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
#defining the passing score as 50 or higher and counting the students with those grades
passing_math_count = school_data_complete[(school_data_complete["maths_score"] >= 50)].count()['maths_score']
passing_math_count
```

```python
#calculating the passing math percentage by dividing and then multiplying by 100 and setting decimal point
passing_math_percentage = round((passing_math_count / total_students) * 100, 6)
passing_math_percentage
```

```python
#defining the passing score as 50 or higher and counting the students with those grades
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >=50)].count()['reading_score']
passing_reading_count
```

```python
#calculating the passing reading percentage by dividing and then multiplying by 100 and setting decimal point
passing_reading_percentage = round((passing_reading_count / total_students) * 100, 6)
passing_reading_percentage
```

```python
#Calculating the overall percentage of students passing in both math & reading
passing_count = school_data_complete[
    (school_data_complete["maths_score"] >=50) & (school_data_complete["reading_score"] >= 50)
].count()["student_name"]
overall_percentage_passing = passing_count /  float(total_students) * 100
overall_percentage_passing
```

```python
#Creating an empty dataframe for the local govt area summary table
local_govt_area_summary = pd.DataFrame()
print(local_govt_area_summary)
```

```python
#Initialize data by creating a list and columns
data = {'Total Schools': [total_schools], 'Total Students': [total_students], 'Total Budget': [total_budget],
        'Average Maths Score': [average_maths_score], 'Average Reading Score': [average_reading_score],
        '% Passing Maths': [passing_math_percentage], '% Passing Reading': [passing_reading_percentage],
        '% Overall Passing': [overall_percentage_passing]}
local_govt_area_summary = pd.DataFrame(data)
local_govt_area_summary
```

```python
#map formatting; converting total students, total budget columns to a float
local_govt_area_summary["Total Students"] = local_govt_area_summary["Total Students"].map("{:,}".format)
local_govt_area_summary["Total Budget"] = local_govt_area_summary["Total Budget"].astype(float).map("${:,.1f}".format)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create bins in which the data for the spending is held
spending_bins = [0, 585, 630, 645, 680]

#Create the group names for the bins
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_spending_df = per_school_summary.copy()

#Use pd.cut to categorise spending based on the bins.
school_spending_df['Spending Ranges (Per Student)'] = pd.cut(school_summary_student_budget, bins=spending_bins, labels=labels, right = False)
school_spending_df
```

```python
#using the following code provided to then calculate mean scores per spending range. Also using the round function and 2 decimal places.
spending_math_scores = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean(),2)
spending_reading_scores = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean(),2)
spending_passing_math = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Maths"].mean(),2)
spending_passing_reading = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean(),2)
overall_passing_spending = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean(),2)
```

```python
#Assembling the new data frame.
spending_summary = pd.DataFrame({"Average Maths Score": spending_math_scores,
                                        "Averavge Reading Score": spending_reading_scores,
                                        "% Passing Maths": spending_passing_math,
                                        "% Passing Reading": spending_passing_reading,
                                        "% Overall Passing": overall_passing_spending})
spending_summary
```
